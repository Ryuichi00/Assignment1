```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import requests
import urllib.request
```

```python
keywords = ['Car', 'Truck', 'Bicycle', 'Boat', 'Airplane']

for keyword in keywords:
    driver = webdriver.Chrome(executable_path = 'C:\ML\chromedriver')
    site = 'https://google.com/search?q='+ keyword +'&tbm=isch'+'&tbs=isz:lt,islt:qsvga'
    driver.get(site)
    for _ in range(0, 10):
        driver.execute_script("window.scrollBy(0,document.body.scrollHeight)")
        try:
            driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div/div[2]/div[2]/input').click()
        except Exception as e:
            pass
        time.sleep(3)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    img_tags = soup.find_all('img', class_ = 'rg_i')
    
    with OSFS('./download') as myfs:
        if (not myfs.exists(keyword)):
            myfs.makedir(keyword)
            
    count = 1
    for i in img_tags:
        try:
            download_path = 'download/' + keyword + '/' + keyword + str(count) + '.jpg'
            urllib.request.urlretrieve(i['src'], download_path)
            print("Number of " + keyword + " images downloaded = " + str(count))
            count += 1
        except Exception as e:
            pass
```
